```python
import folium
import pandas
from folium.plugins import MarkerCluster
```

```python
UNITED_STATES_COORDINATES = (37, -96)
df = pandas.read_csv('coordinates.csv', index_col=0 )
wingstop_map = folium.Map(location=UNITED_STATES_COORDINATES, zoom_start=4)
marker_cluster = MarkerCluster().add_to(wingstop_map)
```

```python
# Create map markers in folium map
for index in list(df.index):
    folium.Marker(location=[df.iloc[index]['latitude'], df.iloc[index]['longitude']]).add_to(marker_cluster)
```
